4 - Given three number x, y, and z,


       Output: x to the power of z,
               x to the power of y,
               x minus y,
               remainder value when x is divided by z

```python
x = 3
y = 5
z = 2

power_z = x ** z
power_y = x ** y
subtract_y = x - y
remainder_x_z = x % z

print("x to the power of z:", power_z)
print("x to the power of y:", power_y)
print("x minus y:", subtract_y)
print("Remainder when x is divided by z:", remainder_x_z)
```

5 - Add

        4 quarters (0.25),  3 dimes (0.10), 2 nickels (0.5) and 1 cent
        Output:  the total amount as dollars and cents.

```python
quarter = 0.25
dime = 0.10
nickel = 0.05
cent = 0.01

quarters = 4
dimes = 3
nickels = 2
cents = 1

total_amount = (quarters * quarter) + (dimes * dime) + (nickels * nickel) + (cents * cent)
dollars = int(total_amount)
cents = int((total_amount - dollars) * 100)
print("The total amount is:", dollars, "dollars and", cents, "cents.")
```

15 - Compare two number a =150, b = 225 and print if they are:

        equal,
        greater,
        greater than or equal to,
        less,
        less than or equal to,  
        not equal
       
        example : a < b
                  print("a is less than b")

```python
a = 4
b = 6

if a==b:
  print("a is equal to b")
elif a>b:
  print("a is greater than b")
elif a>=b:
  print("a is greater than or equal to b")
elif a<b:
  print("a is less than b")
elif a<=b:
  print("a is less than or equal to b")
else:
  print("a is not equal to b")
```
