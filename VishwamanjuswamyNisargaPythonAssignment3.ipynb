Write a Python function that accepts a string and calculates the number of upper case letters and lower case letters.

    Sample String : 'Hello Students, how are you this fine Tuesday?'
    Expected Output :
    No. of Upper case characters : 3
    No. of Lower case Characters : 34

Note: You can use tring methods **.isupper()** and **.islower()**


```python
def upper_lower(s):
    #Write code here!
    uc = 0
    lc = 0

    for st in s:
      if st.isupper():
        uc+=1
      elif st.islower():
        lc+=1
    return uc, lc

    print('No. of Upper case characters : {uc}')
    print('No. of Lower case characters : {lc}')
```

```python
s = 'Hello Students, how are you this fine Tuesday?'
upper_lower(s)
```

**Write a Python function to check whether a string is pangram or not. (Assume the string passed in does not have any punctuation)**

    Note : Pangrams are words or sentences containing every letter of the alphabet at least once.
    For example : "The quick brown fox jumps over the lazy dog"


```python
import string

def ispangram(st):
  #Write code here!
  alphabet = set(string.ascii_lowercase)
  return set(st.lower()) >= alphabet
```

    Weekday Tolls

    Before 7:00 am ($1.15)
    7:00 am to 9:59 am ($2.95)
    10:00 am to 2:59 pm ($1.90)
    3:00 pm to 7:59 pm ($3.95)
    Starting 8:00 pm ($1.40)
    
    Weekend Tolls

    Before 7:00 am ($1.05)
    7:00 am to 7:59 pm ($2.15)
    Starting 8:00 pm ($1.10)

    input : toll_fee(8, True, False)
    Output : $ 2.95


```python
#define your code here

def toll_fee(hour, morning, weekend):
    if weekend:
        if hour < 7:
            return 1.05
        elif 7 <= hour < 20:
            return 2.15
        else:
            return 1.10
    else:
        if morning:
            if hour < 7:
                return 1.15
            elif 7 <= hour < 10:
                return 2.95
            elif 10 <= hour < 15:
                return 1.90
            elif 15 <= hour < 20:
                return 3.95
            else:
                return 1.40
        else:
            if hour < 7:
                return 1.15
            elif 7 <= hour < 20:
                return 2.95
            else:
                return 1.40
```

```python
#call your function

fee = toll_fee(8, True, False)
print("The fee of the hour: ", fee)
```

27 - Write a program that takes a simple password and makes it stronger by replacing certain characters as per list below and appending "!" to the end of the input string.

        i becomes 1
        a becomes @
        m becomes M
        B becomes 8
        s becomes $
        z becomes 2
        o becomes 0
        
        If the input is: mypassword
        the output is: Myp@$$w0rd!

```python
def st_password(pwd):
  rep = {'i': '1', 'a': '@', 'm': 'M', 'B': '8', 's': '$', 'z': '2', 'o': '0'}
  st_password = ' '
  for ch in pwd:
    if ch.lower() in rep:
      st_password += rep[ch.lower()]
    else:
      st_password += ch
  return st_password + '!'


st_password('mypassword')
```
