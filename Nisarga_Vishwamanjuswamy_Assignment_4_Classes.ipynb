Build the class and it as "Purchase" with the following requirements:

    Attributes :
    item_name (string)
    item_price (int)
    item_quantity (int)
    
    Default constructor: With the following default values
    Item_name = "none",
    item_price = 0,
    item_quantity = 0

    Method:
    item_cost()

    This method should calculate the item cost example :
    Coca Cola 10 @ $1 = $10

If there are more than one item then you should add up the cost.
    Item 1
    Enter the item name: Chocolate
    Enter the item price:3
    Enter the item quantity:1

    Item 2
    Enter the item name: Cola
    Enter the item price: 1
    Enter the item quantity: 10

(3) Add the costs of the two items together and output the total cost.

    Ex:
    TOTAL COST
    Chocolate 1 @ $3 = $3
    Cola 10 @ $1 = $10
    Total: $13

```python
class Purchase():
  def __init__(self, item_name="none", item_price=0, item_quantity=0):
    self.item_name = item_name
    self.item_price = item_price
    self.item_quantity = item_quantity

  def item_cost(self):
    return self.item_price * self.item_quantity
```

```python
if __name__ == "__main__":
  purchased_items = []
  total_cost = 0

  num_items = int(input("Enter the number of items: "))
  for i in range(1, num_items+1):
    print(f"Item {i}")
    name = input("Enter the item name: ")
    price = int(input("Enter the item price: "))
    quantity = int(input("Enter the item quantity: "))
    purchased_items.append(Purchase(name, price, quantity))

  print("\nTOTAL COST")
  for purchase in purchased_items:
    cost = purchase.item_cost()
    total_cost += cost
    print(f"{purchase.item_name} {purchase.item_quantity} @ ${purchase.item_price} = ${cost}")

print(f"Total: ${total_cost}")
```

Write a class called Calculator that performs basic functions of a calculator: add, subtract, multiply, divide, and clear.



```python
class Calculator:
  def __init__(self):
    self.res = 0

  def addition(self, *args):
    for num in args:
      self.res += num
    return self.res

  def substraction(self, *args):
    if len(args) < 1:
      return 'invalid entry'
    self.res = args[0]
    for num in args[1:]:
      self.res -= num
    return self.res

  def multiplication(self, *args):
    if len(args) < 1:
      return 'invalid entry'
    self.res = 1
    for num in args:
      self.res *= num
    return self.res

  def div(self, *args):
    if len(args) < 1:
      return 'invalid entry'

    for num in args:
      if num == 0:
        return 'invalid entry'
      self.res /= num
    return self.res

  def clear(self):
    self.res = 0
    return self.res
```

```python
cal = Calculator()
print(cal.addition(5,10))
print(cal.substraction(15,20))
print(cal.multiplication(10,9))
print(cal.div(5,8))
print(cal.clear())
```

Ex. If a new Student object is created, the default output is:

```python
class Student():
  def __init__(self):
    self.name = "DefaultName"
    self.gpa = 1.0

  def set_name(self, name):
    self.name = name
  def get_name(self):
    return self.name

  def set_gpa(self, gpa):
    self.gpa = gpa
  def get_gpa(self):
    return self.gpa
```

```python
student = Student()
student.set_name("Frank")
student.set_gpa(3.7)
print(f"{student.get_name()}/{student.get_gpa()}")
```
