25 - Giving change to customer: Total change amount as an integer input, and output the change using the fewest coins,


      If the input is: 45
      The output is:
                    1 Quarter
                    2 Dimes

      Input: 156
      The output is:
                    1 Dollar
                    2 Quarters
                    1 Nickel
                    1 Penny

```python
def calculate_change(total_change):
    coins = [25, 10, 5, 1]
    change_count = [0] * len(coins)

    for i, coin in enumerate(coins):
        change_count[i] = total_change // coin
        total_change %= coin

    return change_count

def print_change(change_count):
    coin_names = ["Quarters", "Dimes", "Nickels", "Pennies"]

    print("The Output is:")
    for count, coin in zip(change_count, coin_names):
        if count > 0:
            print(f"{count} {coin}")

total_change = int(input("Enter total change amount: "))
change_count = calculate_change(total_change)
print_change(change_count)
```

26 - Write a program that takes in a year and determines whether that year is a leap year.

      Ex: If the input is:  1712
          the output is: 1712 - leap year

          Note : 1) The year must be divisible by 4
                  2) If the year is a century year (1700, 1800, etc.), the year must be evenly divisible by 400

```python
def leap_year(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

year = int(input("Enter a year: "))

if leap_year(year):
    print(f"the output is :{year} - leap year.")
else:
    print(f"the output is :{year} - not a leap year.")
```

27 - Write a program that takes a simple password and makes it stronger by replacing certain characters as per list below and appending "!" to the end of the input string.

        i becomes 1
        a becomes @
        m becomes M
        B becomes 8
        s becomes $
        z becomes 2
        o becomes 0
        
        If the input is: mypassword
        the output is: Myp@$$w0rd!

```python
def strengthen_password(password):
    replacements = {
        'i': '1',
        'a': '@',
        'm': 'M',
        'B': '8',
        's': '$',
        'z': '2',
        'o': '0'
    }
    for char, replacement in replacements.items():
        password = password.replace(char, replacement)
    password += "!"
    return password

password = input("Enter a simple password: ")
strengthen_password(password)
```

28 - Write a program to get 10 multiples of anynumber given.

    Example: if Number = 5
    Output :5,10,15,20,25,30,35,40,45,50





```python
def get_multiples(number):
    multiples = []
    for i in range(1, 11):
        multiples.append(number * i)
    return multiples

num = int(input("Enter a number: "))
result = get_multiples(num)

print(f"Output: {result}")
```
